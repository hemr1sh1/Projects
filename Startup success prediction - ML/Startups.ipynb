```python
x=ds.iloc[:,1:-1].values
x
```

```python
y=ds.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)
```

```python
print(len(xtrain),len(ytrain),len(xtest),len(ytest))
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
```

```python
pred=lr.predict(xtest)
for i in range(len(ytest)):
  print(ytest[i],pred[i])
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
ac=accuracy_score(ytest,pred)
ac
```

```python
cm=confusion_matrix(ytest,pred)
cm
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
confusion_matrix=metrics.confusion_matrix(ytest,pred)
cm_dis=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=[False, True])
cm_dis.plot()
plt.show
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(xtrain,ytrain)
```

```python
predk=knn.predict(xtest)
for i in range(len(ytest)):
  print(ytest[i],predk[i])
```

```python
from sklearn.metrics import confusion_matrix
ack=accuracy_score(ytest,predk)
ack
```

```python
cmk=confusion_matrix(ytest,predk)
cmk
```

```python
confusion_matrix=metrics.confusion_matrix(ytest,predk)
cmk_dis=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=[False, True])
cmk_dis.plot()
plt.show
```
